```python
main_nlp = spacy.load('en_core_web_lg')
```

```python
company_nlp = spacy.load('Company_model')
education_nlp = spacy.load('Education_model')
job_nlp = spacy.load('Jobs_model')
skill_nlp = spacy.load('Skill_model')
```

```python
skill_nlp.pipe_names
```

```python
skill_ner = skill_nlp.get_pipe('ner')
company_ner = company_nlp.get_pipe('ner')
job_ner = job_nlp.get_pipe('ner')
education_ner = education_nlp.get_pipe('ner')

```

```python
new_vocab = ['JOBS','SKILL','COMPANY','EDUCATION']
for item in new_vocab:
    main_nlp.vocab.strings.add(item)
```

```python
main_nlp.add_pipe('ner', source = skill_nlp, name='skill_ner', before='ner')
main_nlp.add_pipe('ner' ,source = company_nlp, name = 'company_ner',after= 'skill_ner')
main_nlp.add_pipe('ner' ,source = job_nlp, name = 'job_ner',after= 'company_ner')
main_nlp.add_pipe('ner' ,source = education_nlp, name = 'education_ner',after= 'job_ner')

 
```

```python
def only_needed(doc):
    lables = ['JOBS','SKILL','COMPANY','EDUCATION','DATE','ORG']
    ents = []
    l = []
    for ent in doc.ents:
#         print(ent.text,ent.label_)
        if ent.label_ in lables:
            l.append(ent)
    l=tuple(l)
    doc.ents = l
    return (doc)

        
```
